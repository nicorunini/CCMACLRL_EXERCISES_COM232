## 2. Compute the Slope $m$ (5 points)

$$
m = \frac{n(\sum xy) - (\sum x)(\sum y)}{n(\sum x^2) - (\sum x)^2}
$$

$$
m = ?
$$


**Answer**
$$
m = \frac{n(\sum xy) - (\sum x)(\sum y)}{n(\sum x^2) - (\sum x)^2}
$$

$$
m = \frac{5(959) - (15)(305)}{5(55) - (15)^2}
= \frac{4795 - 4575}{275 - 225}
= \frac{220}{50}
= 4.4
$$


## 3: Compute the Intercept $b$ (5 points)

$$
b = \frac{\sum y - m \sum x}{n}
$$

$$
b = ?
$$

**Answer**
$$
b = \frac{\sum y - m\sum x}{n}
= \frac{305 - (4.4)(15)}{5}
= \frac{305 - 66}{5}
= \frac{239}{5}
= 47.8
$$

**Answer**

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
x = np.array([1, 2, 3, 4, 5])
y = np.array([52, 57, 61, 65, 70])

# Regression parameters from calculation
m = 4.4
b = 47.8

# Generate points for regression line
x_line = np.linspace(min(x), max(x)+1, 100)  # extend a bit for clarity
y_line = m * x_line + b

# Predicted point for x = 6
x_pred = 6
y_pred = m * x_pred + b

# Plot scatter points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot regression line
plt.plot(x_line, y_line, color='red', label='Regression Line')

# Highlight the predicted point
plt.scatter(x_pred, y_pred, color='green', marker='*', s=200, label=f'Predicted (x={x_pred}, y={y_pred:.1f})')

# Labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression - Exam Scores vs Hours Studied')
plt.legend()
plt.grid(True)

plt.show()
```
